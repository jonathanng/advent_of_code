```python
class AdventSolution:
    
    def __init__(self, input=None, path=None):
        if input is not None:
            self.input = input
        elif path is not None:
            self.input = self.get_input(path=path)
        
    def get_input(self, path):
        with open(path, 'r') as f:
            return ''.join(f.readlines()).split('\n')
        
```

```python
class Day2(AdventSolution):
    
    def _iter_line(self):
        
        for line in self.input:
            dimensions = line.split('x')
            l, w, h = [int(d) for d in dimensions]
            yield l, w, h
    
    def required_wrapping_paper(self):
        
        total = 0
        
        for l, w, h in self._iter_line():
            sides = [l * w, w * h, h * l]
            total += 2 * sum(sides)
            total += min(sides)

        return total
    
    def required_ribbon(self):
        
        total = 0
        
        for l, w, h in self._iter_line():
            dimensions = sorted([l, w, h])
            total += 2 * (dimensions[0] + dimensions[1])
            total += l * w * h
                
        return total
      
assert Day2(input=['2x3x4']).required_wrapping_paper() == 58
assert Day2(input=['1x1x10']).required_wrapping_paper() == 43
Day2(path='day_2_input.txt').required_wrapping_paper()

assert Day2(input=['2x3x4']).required_ribbon() == 34
assert Day2(input=['1x1x10']).required_ribbon() == 14
Day2(path='day_2_input.txt').required_ribbon()
```

```python
input = '^><^>>>^<^v<v^^vv^><<^<><<vv^<>^<^v>^vv<>v><vv^^<>>^^^v<<vv><<^>^<^v<^>^v><<<v^<v<<<v<<vv<v<^><^>><>v>v^<<v^^<^v<><^>^<<^^^>v>>v^^<v>>^>vv><v>>^>>v^>^v>^<^^v>^>^^v<v>^^<v<>>v^^v><^><^<<>v^<^<^v<v>v^>>>v^v^>^<>^v<^^vv<v>^>^<>^^<vv^<><<v<^<^^>vv<>^>v<^>^v>v^>^v<>^><>><vv<>v^v<><>v^v>>>>v^^>^><^^<v<^><^<v>>^v^v<>v<<<^<<vvvv<<v^vv^>v^^^<^^^<v>>v<^v>>>>>v<^^^^>v<^<><v>>>>><v>>v^vvvv^^<v^<>^v<^v^>v><^>^v<<>>vv^>v>v^^>vv^<^vvv<>><>><><^^^<v<>^<^^^<v><^v>>v>^v<v^vv^<>^^^>v^^^v>>^v^^<^>>^>^<<v>>>^^<>>^vv>v^<^>>>><v<><><^^v<><<<<^^<>>^<vvv^><>v<v<<<<><v<<v>v<v^><vv<v^>^<^>v^^><^v>^^>v<>^v^<>^vv^><v^^vv>vvv>v>^<vv^>>^>>^>><>>>^^^^v<vv>^<>v^^><v^>^<>v<^^v><v<<><^v><>^^^^^v^v>>^^v><<><<vv>^^^^><^>v>><<<^v>v^^>^v^<^^v>v<^<<>>^v<<<v<<>>v<^v^><vv<v^v>v^<v>><v>^v<<<vv^>v<v>>v>>v><v><v^>v^^v>^v^>>>><>^>v>^v^>>>>v^<<vv<^v><<>v<v^<^^<<v<^v^^v^>vv><vv<v^<^>><^^>^<><^^<v<><^v^v^<^^>^<v><^<v>v^<<<^^v<v>^v>>><>^^>vv<<^v^<<<<^^>>>v>v<<<>^^>>>v>^>v>vv<<>^<^><v^>^^<^<v<<v<^>>^v^<vvv><>v^><<v>^^<v^vv^^^<vvv^<^>^>vv>><^v<^<<v<><<><<^^<><><vv>v>^<v>>^<>>^^v>vv^<^^v>><^vv^<<v^^><<>vv<v<><v<><v^^^v^v>^v<^<>v^^>><>^<^<v^<v^v^>v<<<^<<^>>>^^<^^v>v^<v>vvvv>v<>><^>^<<<<v^<v<>v^^^v<>v>^<v<<^^v^^<>^<<v^^<^<v>v>>v>>v^>^<vv<<<<<^<><>v><>>>v^>^v<^<><<v<^v^^<^<><^>^^^>^><>^><<vv>^<>vv<<v^v<<<<<>>>v<vv>^v>^>^>^<^><>v<><>>>^^<v>^<^v>>^<><v^><v^>>>v<v^^vvv^><v<v>v^>vvvv>>><^>v<>^^^>v>>v^<v<>v^>^<v^>^<<^>^>>v<<><<v^^>>v^<v^<^v^>^>v^><<^<v>v^<v>>^^<<v>v><<<^v^<>^<>^>>^<<v>^^<>^v<>v^>>><<v>><v^>^><v^<><v><>><v^<>vv>v^<^^^>v>^^<vv>>^v<><>>><>><^<>>v>v^^>^^<^^>^>>v>vv^^v<^<^v><vv<v<^>><<vvv<<><^>^v>^^^<<>v^<v<v><<v>^^v<<<>^^vv<^>vv>^>^<><<>vv<^>v^vv>^^^v><<^vv>^v<><v^^^^v^>vv^^<^<>^^v^<^vv<v<vv<>v>v^^<>^^>^^>^<><<^v>^><^^vvvv<><>^<v^^>v<>^><>v>><>vv^<<><<>><>v<^>^v>>^^v><<<>>^<^v^<v<<<v^>^^<^<><><^><<<<^<vv><v<<><vvv^^><vv>^<<vv<<<^v<>>><><>>v><<<v>vvvv^^vv<v>><<^v^vvv><><vv>v><>v<<<^<v^>><^^>v^<v>><v>^^^v^v>><<<v<^^>>^v<>v^<vv^^<<v<v>v<<<<^^^v^v<<>>>v>>vv>^^<><^v<v><>>v^>>>>>^>v^v^<^v^v^vvv>v<v<^>vv^<<v>vv>>v^^vv<^v>>>>vv<>v<>^^vv^<v>v^>>vvv<<<v<<^vv^^^^>v>v>^><<<^>v^><v<^<<<v>^v^^^><<><<<^^<^^<>^<v>^<v<<v<^^vv>v<^v><v><v<>^v<^<v<^<v^v><v>><v<v<<>^<v<>>><>^v^v<<^><v^<<v<v^>^>v><^>^vv^^<v<v<vv<v>^v^v^>^<<>>>>>v^<>^>v^vv^><<>>^^<>v^><v>^vvv^>v^v><>^><<>v>v<^<^><^^vv<<><>>v>>v><vv>>^v<<>^vv<>^vv>v>v>^>^>>><><<>v<v>^<<^v^^<<<><v>>vv<^<vv<vv^<<v<<^v><<>v<^^^<<^v^>^v>^^^v^v>>>v>v^v>^>^vv<^^<<vv^>^<<<vv>v^<><<^vvv^^><>vv^v>v>^><<^^^^vvv^<vvv>><^v<^>^<>>^<v<<vv>>><v>vv^<>><v^<v>^v>^>v>^<^<^^^<<vvvv^>>>>>>>v><vv>^<>^^v^><>><^v^^<v^v<<<<v^>><>v^v<vv<><^<<<<^>^^>vv>><^v<v^v<<>^vvv>v^^><^^<^<>>^^v^vv<>v<^<<<v^^^><v<vv<<>v>v<>^v^><v^vv^v^^v<^^v^^v><>v<^v>><<^<^v^>><<vv<<^>^<<v^<>^><>v><vv^v>>^<v<<<^>vv<^v>^>v<<v>^>>^>>v^<v<v>>^v<^v^v><<><>^><<<><v<vvvv<v^<v^v><>^<>^^^^v>^>^vvvvv>v>>v><<vv<<v<><<^><<^v><<v<<<v><vv<^>^v>>>>^v<^v<<>>^>^<<vv^<^>v>><<^>^>^v><><>^><<v<>v^><<^v^<^^><^^v^<<^v^^>>^v^<^><vv>v^^<<^^^<><>^>v^v>v^>^v^vv>^^>>>>^^<^>>>^^v<vv<><^^<vvv<^^^vv>v<v<v>><<<>^>^^>^>^v<<<<>>^<<>><v>>v>^^<^v<>v<>v^>v^><^<^^><v^^v>^^vv<v<<>><<vv<>>v>^<<<<v<<v>^><^^<^<^<v^<<^^v>^v<^>v^v^<v^vv^>^^><^>v^v>>^^v^><vv<v<v<v>>>>><<><v><v^v^<v^<^^<v<>^>v>v<>>>v>^^^^>><v^v^^v<<<>v^<<^<v>>>><^v^<<><v<>>v><><v<v^v>^v^^<v<^<^^v>><<vv<<vv><>>^>^>vv<^<>^vvv^v<v^^<>v^v>^^<<<<<>^v^>^<>v^^<>v^v<vv>^<>vv^<^vv>><v^^vvvvv>><<>v<vv^<^<vv^v^<>^^<v^<vv^<v^v^v<<^>^>^>^^>>>vvv>^>v>v>>>^>vv^><>^><>v>^^<v^>^><<v>><<<>>v<vvvv^>^v<^<>^<v>^<>^^<<><>^v<><>>>^vv<^<<^<^v>v<<<<<^^v<^v<><v<<><^>v>^v>>^v^><^^^^v<><><>vv^<>vv<^v<^^><v^<^><^^v^v^<^^<<><v>v<v<v^<<^v><>v^v<^>vvv><<^v>>v><><v<<^>>>v<^>>v>^<>><>^<v^v^<vv<<^>v<^^>^<^v<^<<^^v<>>^>^>^v^^v^v<v^^vv^<v>>v><vv^vv>v<>v^>v^^>^^>><v><v^<<><<>><<^^>><^v<v<><<><<><v<v^<^<v>>>><v^^v^^>>>^^^^^<<vv<^><>^<<<vv^^^>^><<<v<^v>^<v<^>^vvv<<>vv><<>v>v^v>>>>>^<>><^^^><<<<v><<vv>>>v<^<vv^v^<<v>>>>^^vvv>v<>><v>>>v>>^v^vvv<<>vvv<<^^^<>vv^^v<<>^^^>>^<^v^<^^>v^><v>>^<<^v<<vv<vv>v^>>^>v^><^><>^>>>vv>><^^^>vv<<^^vv><^<>^>^^<^<>>^vv^>>^v><>v^>>><<<^^<^>^>v<^>^<^^<>>><^^<>^v^<<vvv<v><>vvv><v>v^v<<^<v>^^><<^vv^v>v>v<<^v^<<<>^><><vvv>v>^vv^v<>vv^>^^<^>^>v^^<vv^>v><v<<<><>>^v<^<><><^<v^^<<^<v>vv<><<>v^<v^>^>^^<><<>^<^<<v^^v<v^<><<>v>><^<<>^>^v^v<v^v><^>>^v<^>v<<>^^^<^v>>>^<v>vvvv<<v^<^^>vvvv>v<>v<v><vvvvv>^<><>vvv<>^<<>^>>>>v^<^<><^v>v^>>v><>^><<v^>^<<>^>^v^<v^^>>^v><v>^<v><>v^<^^>v>^>>>v^v>>>^<>^<>>>>>v>>vv^v<><<<><><v><<vv<<v<><>>vv<^<vv>^v<<>v^v<^v<><v>>^v>>vvv^^v>>v>^>^>v><v><^>^^<<>^v<^<<<<^>v<^>>v^<^v>^v<<>^>^vvv<^^vv>^vv>vv<>>v>v<v>>v^<<<<<^^v^>v>^<<<v^v>>v<v><vvv><v>^<vv><<>>^<^>^^<>>>>^<^v<>v^^>^<^^v<^><>><v>>^v^vv<^v<^><<vvv<>><>><^^>^<^v^<^<>v<<<^v>v^^^<>v^<v^>^v^>><>^^<v<^><<^^v^<>^<^vv>>><^v><v^>vv<^v<<<v^>>v>v^v>^<v>v<^<>v^vvv>^vv<<<<v><^><v>>^^>><^v><<^>v^^<<v^^<^<><<<<>^<v<^v^>v<<^^>v<<<<<vvv<v<^>^>^>^>>^>>>v^<<v>>^^v><vv<^v<v<^^^>>>^vvv<^v<>>>vv>^^><^v>vv^>>v>v^<>^<vv>^>^<<^>^^^>>^vv>^^>vvvv<>>^^^^>>>v>v^^>vv>vv^<<>^><^<v^vvvv><v<><v>><<<v<v<<^v><vv^vv^<>>>^>^<v<^v<>><^<vv^^><v>v^>v^<><v^vvv>^>v^^v^>^^>v<<<<^<<^>>v>v^^^<<<v>>>^^v>v<v><<<<^^^v>^vv^>><>^v<v<<^^<<<<><>>>v>vvv^v^^v^>>vv>^>><>^v><^v^><^^>vv>^<^<^>><v>v>><><><v>^>^>v>vv>vv>^^>v>v^><v<<v^<>^>^v>^^v>^<^v<>>vvv^^>^>vv<v<v<<^<^<v^<>v^^v<^<^>vv^^<v><^^^>v>vv<<v>v<<v^<v^^><vv>^>^v^<^>v<^>^<>vv^><v<^><>>^>>^<^><<>^<^>v>v><>>>^<<^><<v><^v<v><>>vv<^><v^>>v>v>>>>^^>v<^v^>><<^<>>v><^><<^>^<vv^^<><<>><vvvv^>^^<><^^v>^^>vv>^v<v>>^^v^<v<^><^<<>>v^^^<^><^<<><<v<>><<>^v>vvv^vvv^^>>^<^<v>><>^<<<<^^<>>>v^<<^^v>><><<v<^>v>^v<v^>v>vv^><>^><<><^^>^>^<><>><^^<v^v<^><><><v>^<v<<v^<<^^^v<v<^v<>>><^v<<<<>>^v>^^vv^v^<<v>><<<v>vv>>v>>^v^<>>vv^<^>^<<>v<<<^vv<^vv^vv<^v^^^<vv^>v>>v<^^<^^vvv<^^v<>>>^>v^><v>^^><>vv>v>v<<<^^v<^vv^v>^^^>>>^^<>^^<^vvv>><><<><^<v>><<>^>^^<v^v^>vv>vv<v>^^<^^<<><><<v><v^^>v><v><<>v>vvv<^^^^<^>>><<<^^^<^>vv^^v>>v<<v^^<vv^<^>vvv^^v^^<^<vv>v<^<>^<<vv^^>^v>>^><><>v<v<v<>><v>>>^^>>v^><v^^<^>><>v<><<v^v<v<<>>>><>>>>><<^vvv<<><><<>^><><<^^v><<^>v>^>^v>v>>^^<><^>vv<^<^v>v<><^<<v<><^><>^^^<v^<><vvv^^^<>^^v><v<<<v>><>^>^vv<v^<vv>v>v^vv<v^v<v>^v^>v><>v^><>v>^^^^><<vv^><v<<v<^<>^v^^^>^^><<<v<^<v^>^^>v><vvvvv^<^<v^^>v<^v^^vv^<<<<v><^>v>v^v><><v^<<^<<v<^^^>^><v^v^<><><>^v<v>^<>^v>^v>v^<><^><v>>v<<^><^vv^<><^<>><>><v<v><<^^^^>v<^<^vv<><^vv><<^<<v>v^>>^v>^>v^^v>vv<v>v<<v>v<>^>>vv^>>><>^v^^<^>v<<^<^^v^^v^<<v<<v<^v<>vv^<v>><^v<^>>>vv^^<v^<>^^v<v<v>>^><^^^<><<^^>v<<vv>><<vvv>><<v^v^>><>vv^><<^>^><^v<^<^<vv<^^vv>v^v<<<<<<><<vv^vv>vv>v<^><<><><<>>v>><v><^>^v>^v^<>v^^^><^^<<<^vv^vv>^v^vvv^^>v^<v>><^<^<^<>^vv<vv^v^^>^^^>vv^v>>><<<^<>>v>v<^^<><v>>><><^v^^<<><<<>^<^^v^>v<vv^^^^>><v><^<<v<<v<>^>^>>^<>^v><>>^<v<vv^<<^<<>vv^>^^<<<^v<>>^v<>vvv<<^^<<><vvvvv<<^<^^<>>>>^^<><>^><>^v<v^^v<<v^^<^<^>v<v>^v<^>^v<>v^vv<><<v>^vvv<><<^>>^^><><>^<>^>v^^v^><v<><>>v><v^<v<<v>><^v>^<v<^>v<<<>vvv^<^^v<vvv^vv<>^<>^>>v<>^^><><v>>^><^^vv>><<>><v><^><>>^vv>v<vv<>v^v^^v<<^^<vv>v^^vv<<^<<><>^<><v^><^<^<>>^vv<v>v>>^<^vv>^vv^>v>^<><^><^<>v^v^^<^<>^^v>>><^v<>v^v<<^>v><>^^<<v^v<>v^>>v>^<><vv^v<v^<vv<>^>^>^<^>v><<><><><<<>^>><v^^><^>><v>>^v<<<^<<>^><<^>>>>>v<^>v>>v^<v^>^>v^^><>v^v^vvvv<v<v<>v>>><<>^<<vvv><v^v^>v<v^^^>>^<v>>^vv^^<vv><^>>v<v^><vvv<^^>>vv^v<^<>^v^<<v>^<<><<<^vvv^>^^<<>>><v<^>vv<<^<><^v<^<><<^^>vv^v>v^^^>>>>^>vv<<v>v>>^^v^^><>v<<^><^<v^>>^>v^v>><^v^>v<<^<v><^<^<^<>>v^^>><<<>v<v>v<^^>^vv<<<^^<v<>v^^>v<<><^<>^^>^v<>v>><^^^vv^>^><>v^^<v^<>>^<v^^^><v<><vvv>v>^<<^v>^>>>>><^^^<>v<v>>v^^<^v^>>v^<<v^>^>v^v>>>>^>>vv<>^<^v><v^^<>v>v^v>^<>^>v<vv><<v<^v<<^v<<^v^vv<><>^<>>^<>>^<>v^><<>^v>>^^^^<<^v><>^<^>^^v><^^<^<v^<^^v>^v><vv>v<<^>^>><<^^^vvv<<^vv<^^>v^^vv^<^^<<^^>>^^<vv<v<<v^^<<v<^vvv<<><<v>v^>>v^^>v<^>^><v<^>v<v^v<v^^<>v>><<v^v^v<^^^><v>v><^<^vv>^^v>^>v<<^vv><^^^^^^><<^>>>^v<>^^v<<<>><<<v^><>^<<<v>v^>^^^<^><v>^^^v<<>v<v>^<v^>><<^^<<^v<<>^v>>vv>><v<^><v<<<vvv><vv><<^v^^<v^vvv<^v>>v^v<v^v^>>^^v<><^^^<^^>v>^<><v<<v^^>vvv^v^^<v<v^v>^>v^^v<^><v^^<<<<>^^>>^v<><^><^<<^vv^<><<>v^vv^<v^<><<<^^>v<<>>>v<>v<><<<v>^v>^^v>^^>v>^>^>v<>><>^>^>^vvvv<^<v^<>^^^^v>v>><<v>>^<vv>>^<v<^v^vv>><>^^>v^^<<><^<v>><<<<>v>^^><v^^v<<v<><vv^v>^<v^^>v<<<<v^v<<>>vv<v<<<v>v>>v<^v>>v>v^<<<>^>^>^<>v<^^vv><^v<<^v<vvv^vv>v<^<<^^vv^^>vv<^>v>^^<<v^<<^^v<>^>v<<^^<^>^^^v^^<v<^<^>>>v^vv^<^v>^<>^<^<v<^v>>>^<^v<><v<^vv<v>v><v^v^^v<vv><^^<><>^>v<^<^vv>><^v><v<>^<>^^>^<><<<v^>>^<>><<><v>vvv^<<^<vv<v><v<^<<<^>^>>v<^>>vv>^v^^^v<>v<>><>^vv^>vv^'

class Day3(AdventSolution):
    
    INSTRUCTIONS = {
        '<': (-1, 0),
        '>': (1, 0),
        '^': (0, 1),
        'v': (0, -1),
    }
    
    def calc_houses_visited_count(self):

        x, y = 0, 0
        
        visited = collections.defaultdict(int)
        visited[(x, y)] += 1

        for s in self.input:
            x_delta, y_delta = self.INSTRUCTIONS[s]
            x += x_delta
            y += y_delta
            
            visited[(x, y)] += 1
            
        return len(visited)
    
    def calc_houses_visited_count_robo(self):

        x, y           = 0, 0
        x_robo, y_robo = 0, 0
        
        visited = collections.defaultdict(int)
        visited[(x, y)]           += 1
        visited[(x_robo, y_robo)] += 1
        
        is_robo = False

        for s in self.input:

            x_delta, y_delta = self.INSTRUCTIONS[s]
            
            if is_robo:
                x_robo += x_delta
                y_robo += y_delta
                visited[(x_robo, y_robo)] += 1
                # print(f'robo {x_robo=}, {y_robo=}')
                is_robo = False
            else:
                x += x_delta
                y += y_delta
            
                visited[(x, y)] += 1
                # print(f'{x=}, {y=}')
                is_robo = True
            
        return len(visited)
            
assert Day3(input='>').calc_houses_visited_count() == 2
assert Day3(input='^>v<').calc_houses_visited_count() == 4
assert Day3(input='^v^v^v^v^v').calc_houses_visited_count() == 2
Day3(input=input).calc_houses_visited_count()

assert Day3(input='^v').calc_houses_visited_count_robo() == 3
assert Day3(input='^>v<').calc_houses_visited_count_robo() == 3
assert Day3(input='^v^v^v^v^v').calc_houses_visited_count_robo() == 11
Day3(input=input).calc_houses_visited_count_robo()
```

```python
class Day5(AdventSolution):
    
    def is_nice_string(self, s):
        
        letter_count = collections.defaultdict(int)
        
        for letter in s:
            letter_count[letter] += 1
        
        vowel_count = sum([
            letter_count[vowel]
            for vowel
            in 'aeiou'])
        
        if vowel_count < 3:
            return False
        
        if not self.has_double_letter(s):
            return False
        
        for word in ['ab', 'cd', 'pq', 'xy']:
            if word in s:
                return False
        
        return True
        
    def has_double_letter(self, s):
        
        for x, y in zip(s[:-1], s[1:]):
            if x == y:
                return True
            
        return False
    
    def calc_nice_count(self):
        
        count = 0
        
        for word in self.input:
            if self.is_nice_string(s=word):
                count += 1
                
        return count
    
    def contains_pair(self, s):
        
        last_pair_positions = {}
        
        index = 0
        while index < len(s) - 1:
            
            pair = s[index:index+2]
            
            if pair not in last_pair_positions.keys():
                last_pair_positions[pair] = index
            elif last_pair_positions[pair] <= index - 2:
                # print(f'{last_pair_positions=}')
                return True
                
            index += 1
            
        # print(f'{last_pair_positions=}')
        return False
    
    def is_nice_string_better(self, s):
        
        if not self.contains_pair(s=s):
            return False
        
        index = 0
        while index < len(s) - 2:
            if s[index] == s[index + 2]:
                return True
            else:
                index += 1
            
        return False
        
    def calc_nice_better_count(self):
        
        count = 0
        
        for word in self.input:
            if self.is_nice_string_better(s=word):
                count += 1
                
        return count
                
assert Day5().is_nice_string(s='ugknbfddgicrmopn')
assert Day5().is_nice_string(s='aaa')
assert Day5().is_nice_string(s='jchzalrnumimnmhp') == False
assert Day5().is_nice_string(s='haegwjzuvuyypxyu') == False
assert Day5().is_nice_string(s='dvszwmarrgswjxmb') == False
Day5(path='day_5_input.txt').calc_nice_count()

assert Day5().is_nice_string_better(s='qjhvhtzxzqqjkmpb')
assert Day5().is_nice_string_better(s='xxyxx')
assert Day5().is_nice_string_better(s='uurcxstgmygtbstg') == False
assert Day5().is_nice_string_better(s='ieodomkazucvgmuy') == False
Day5(path='day_5_input.txt').calc_nice_better_count()
```

```python
class Day6(AdventSolution):
    
    def iter_coords(self, line, ignore):
        start, end = line[ignore:].split(' through ')
        start = [int(x) for x in start.split(',')]
        end   = [int(x) for x in end.split(',')]

        for x in range(start[0], end[0] + 1):
            for y in range(start[1], end[1] + 1):
                yield x, y
    
    def follow_instructions(self):
        
        is_on = collections.defaultdict(int)
        
        for line in self.input:
            if line.startswith('turn on'):
                for x, y in self.iter_coords(line=line, ignore=8):
                    is_on[(x,y)] = 1
            elif line.startswith('turn off'):
                for x, y in self.iter_coords(line=line, ignore=9):
                    is_on[(x,y)] = 0
            elif line.startswith('toggle'):
                for x, y in self.iter_coords(line=line, ignore=7):
                    is_on[(x,y)] = (is_on[(x,y)] + 1) % 2
                    
        return sum(is_on.values())
    
    def follow_instructions_fixed(self):
        
        brightness = collections.defaultdict(int)
        
        for line in self.input:
            if line.startswith('turn on'):
                for x, y in self.iter_coords(line=line, ignore=8):
                    brightness[(x,y)] += 1
            elif line.startswith('turn off'):
                for x, y in self.iter_coords(line=line, ignore=9):
                    brightness[(x,y)] = max(0, brightness[(x,y)] - 1)
            elif line.startswith('toggle'):
                for x, y in self.iter_coords(line=line, ignore=7):
                    brightness[(x,y)] += 2
                    
        return sum(brightness.values())
    
Day6(path='day_6_input.txt').follow_instructions()
Day6(path='day_6_input.txt').follow_instructions_fixed()
```

```python
class Day7(AdventSolution):

    wire_signal_instructions = {}

    @functools.lru_cache
    def _calc_wire_signal(self, wire_name):
        try:
            return(int(wire_name))
        except ValueError:            
            wire = self.wire_signal_instructions[wire_name]
            return wire['f'](**wire['kwargs'])
    
    def solve(self):

        for line in self.input:
            
            # print(line)
            
            signal, wire_name = line.split(' -> ')
            
            if ' AND ' in signal:
                x_name, y_name = signal.split(' AND ')
                self.wire_signal_instructions[wire_name] = {
                    'f':      lambda **kwargs: self._calc_wire_signal(wire_name=kwargs['x_name']) & self._calc_wire_signal(wire_name=kwargs['y_name']),
                    'kwargs': {'x_name': x_name, 'y_name': y_name}}
            elif ' OR ' in signal:
                x_name, y_name = signal.split(' OR ')
                self.wire_signal_instructions[wire_name] = {
                    'f':      lambda **kwargs: self._calc_wire_signal(wire_name=kwargs['x_name']) | self._calc_wire_signal(wire_name=kwargs['y_name']),
                    'kwargs': {'x_name': x_name, 'y_name': y_name}}
            elif ' LSHIFT ' in signal:
                x_name, num = signal.split(' LSHIFT ')
                self.wire_signal_instructions[wire_name] = {
                    'f':      lambda **kwargs: self._calc_wire_signal(wire_name=kwargs['x_name']) << kwargs['num'],
                    'kwargs': {'x_name': x_name, 'num': int(num)}}
            elif ' RSHIFT ' in signal:
                x_name, num = signal.split(' RSHIFT ')
                self.wire_signal_instructions[wire_name] = {
                    'f':      lambda **kwargs: self._calc_wire_signal(wire_name=kwargs['x_name']) >> kwargs['num'],
                    'kwargs': {'x_name': x_name, 'num': int(num)}}
            elif signal.startswith('NOT '):
                x_name = signal[4:]
                self.wire_signal_instructions[wire_name] = {
                    'f':      lambda **kwargs: ~self._calc_wire_signal(wire_name=kwargs['x_name']) % 65536,
                    'kwargs': {'x_name': x_name}}
            else:
                x_name = signal
                self.wire_signal_instructions[wire_name] = {
                    'f':      lambda **kwargs: self._calc_wire_signal(wire_name=kwargs['x_name']),
                    'kwargs': {'x_name': x_name}}
                
        return {
            name: self._calc_wire_signal(wire_name=name)
            for name
            in self.wire_signal_instructions.keys()
        }
    
    def solve_rewire(self):
    
        wire_signals = self.solve()
        
        self.wire_signal_instructions['b'] = {
            'f':      lambda **kwargs: self._calc_wire_signal(wire_name=kwargs['x_name']),
            'kwargs': {'x_name': wire_signals['a']}}
        
        self._calc_wire_signal.cache_clear()
    
        return {
            name: self._calc_wire_signal(wire_name=name)
            for name
            in self.wire_signal_instructions.keys()
        }
           
input = '''123 -> x
456 -> y
x AND y -> d
x OR y -> e
x LSHIFT 2 -> f
y RSHIFT 2 -> g
NOT x -> h
NOT y -> i'''

wire_signals = Day7(input=input.split('\n')).solve()
assert wire_signals['d'] == 72
assert wire_signals['e'] == 507
assert wire_signals['f'] == 492
assert wire_signals['g'] == 114
assert wire_signals['h'] == 65412
assert wire_signals['i'] == 65079
assert wire_signals['x'] == 123
assert wire_signals['y'] == 456

Day7(path='day_7_input.txt').solve()['a']
Day7(path='day_7_input.txt').solve_rewire()['a']
```

```python
class Day8(AdventSolution):
    
    def solve(self):
        
        character_count = 0
        in_memory_count = 0
        
        for line in self.input:
            # print(f'{line=}, {len(line)=}')
            character_count += len(line)
            
            index = 1
            while index < len(line) - 1:

                # print(line[index])
                in_memory_count += 1
                
                if line[index] == '\\':
                    
                    if line[index + 1] == 'x':
                        index += 4
                    else:
                        index += 2
                else:
                    index += 1
                    
            # print(f'{in_memory_count=}')
            
        return character_count - in_memory_count
    
    def solve_reverse(self):
        
        new_count = 0
        old_count = 0
        
        for line in self.input:
            # print(f'{line=}, {len(line)=}')
            old_count += len(line)
            
            new_count += 2
            
            index = 0
            while index < len(line):
                if line[index] == '"':
                    new_count += 2
                elif line[index] == '\\':
                    new_count += 2
                else:
                    new_count += 1
                    
                index +=1
                    
        return new_count - old_count
        
input = r'''""
"abc"
"aaa\"aaa"
"\x27"'''.split('\n')
        
assert Day8(input=input).solve() == 12
Day8(path='day_8_input.txt').solve()

assert Day8(input=input).solve_reverse() == 19
Day8(path='day_8_input.txt').solve_reverse()
```

```python
class Day9(AdventSolution):
    
    def calc_route_distances(self):

        distances = {}
        cities    = []
        
        for line in self.input:
            # print(f'{line=}')
            route, distance = line.split(' = ')
            city_1, city_2 = route.split(' to ')
            distances[(city_1, city_2)] = int(distance)
            distances[(city_2, city_1)] = int(distance)
            cities.append(city_1)
            cities.append(city_2)
        
        cities = set(cities)
        
        route_distances = {}
        for route in itertools.permutations(cities, len(cities)):
            
            total_distance = 0
            for index in range(len(cities) - 1):
                distance = distances[route[index], route[index + 1]]
                total_distance += distance
                # print(f'{route[index]=}, {route[index + 1]=}, {distance=}')
                
            route_distances[route] = total_distance
            
        return pd.Series(data=route_distances).sort_values()
    
    def calc_shortest_route(self):
        
        return self.calc_route_distances()[0]
    
    def calc_longest_route(self):
        
        return self.calc_route_distances()[-1]
            

        
input = '''London to Dublin = 464
London to Belfast = 518
Dublin to Belfast = 141'''.split('\n')

assert Day9(input=input).calc_shortest_route() == 605
Day9(path='day_9_input.txt').calc_shortest_route()

assert Day9(input=input).calc_longest_route() == 982
Day9(path='day_9_input.txt').calc_longest_route()
```

```python
class Day10(AdventSolution):
    
    def next_sequence(self, sequence):
        return int(''.join([x + str(len(list(group)))
            for x, group
            in itertools.groupby(iterable=str(sequence))]))
    
    def solve(self, rounds=40):
        sequence = self.input
        for _ in tqdm.tqdm(range(rounds)):
            sequence = self.next_sequence(sequence=sequence)
            print(f'{index=}, {len(str(sequence))=}')
        return len(str(sequence))
    
assert Day10().next_sequence(sequence=1) == 11
assert Day10().next_sequence(sequence=11) == 21
assert Day10().next_sequence(sequence=21) == 1211
assert Day10().next_sequence(sequence=1211) == 111221
assert Day10().next_sequence(sequence=111221) == 312211
assert Day10(input=1321131112).solve(rounds=40) == 492982

Day10(input=1321131112).solve(rounds=50)
```

```python
import string

class Day11(AdventSolution):
    
    def next_password(self, password):
        
        if type(password) == str:
            password = list(password)
        
        password[-1] = chr(ord(password[-1]) + 1)
        if ord(password[-1]) == 123:
            return self.next_password(password[:-1]) + 'a'
        else:
            return ''.join(password)
        
    def is_valid_1(self, password):
        # Passwords must include one increasing straight of at least three letters, like abc, bcd, cde, and so on, up to xyz. They cannot skip letters; abd doesn't count.
        for index in range(0, len(password) - 2):
            if ord(password[index]) == ord(password[index + 1]) - 1 == ord(password[index + 2]) - 2:
                return True
        return False
    
    def is_valid_2(self, password):
        # Passwords may not contain the letters i, o, or l, as these letters can be mistaken for other characters and are therefore confusing.
        for letter in 'iol':
            if letter in password:
                return False
        return True
    
    def is_valid_3(self, password):
        # Passwords must contain at least two different, non-overlapping pairs of letters, like aa, bb, or zz.
        count = 0
        for letter in string.ascii_lowercase:
            if letter *2 in password:
                count += 1
                
        return count >= 2
    
    def is_valid(self, password):
        return (
            self.is_valid_1(password=password)
            and self.is_valid_2(password=password)
            and self.is_valid_3(password=password))
    
    def next_valid_password(self, password):
        while True:
            password = self.next_password(password=password)
            if self.is_valid(password):
                return password
        
    
assert Day11().next_password('xx') == 'xy'
assert Day11().next_password('xy') == 'xz'
assert Day11().next_password('xz') == 'ya'
assert Day11().next_password('ya') == 'yb'
assert Day11().is_valid('hijklmmn') == False
assert Day11().is_valid('abbceffg') == False
assert Day11().is_valid('abbcegjk') == False
assert Day11().next_valid_password('abcdefgh') == 'abcdffaa'
assert Day11().next_valid_password('ghijklmn') == 'ghjaabcc'

Day11().next_valid_password('hepxcrrq')
Day11().next_valid_password('hepxxyzz')
```

```python
class Day12(AdventSolution):
    
    def calc_sum(self, data, ignores=[]):
        # print(f'{data=}')
        if isinstance(data, list) and not isinstance(data, str):
            return sum([
                self.calc_sum(x, ignores=ignores)
                for x
                in data])
        elif isinstance(data, dict):
            for ignore in ignores:
                if ignore in data.values():
                    return 0
                
            return sum([
                self.calc_sum(k, ignores=ignores) + self.calc_sum(v, ignores=ignores)
                for k, v
                in data.items()])
        elif isinstance(data, int):
            return data
        elif isinstance(data, str):
            return 0
        else:
            raise NotImplementedError(data)
        
    def calc_sum_str(self, data, ignores=[]):
        return self.calc_sum(data=eval(data), ignores=ignores)
    
    def calc_sum_input(self, ignores=[]):
        # print(self.input)
        return self.calc_sum_str(data=self.input[0], ignores=ignores)
        
        
        
assert Day12().calc_sum_str('[1,2,3]') == 6
assert Day12().calc_sum_str('{"a":2,"b":4}') == 6
assert Day12().calc_sum_str('[[[3]]]') == 3
assert Day12().calc_sum_str('{"a":{"b":4},"c":-1}') == 3
assert Day12().calc_sum_str('{"a":[-1,1]}') == 0
assert Day12().calc_sum_str('[-1,{"a":1}]') == 0
assert Day12().calc_sum_str('[]') == 0
assert Day12().calc_sum_str('{}') == 0

Day12(path='day_12_input.txt').calc_sum_input()
Day12(path='day_12_input.txt').calc_sum_input(ignores=['red'])
```

```python
class Day13(AdventSolution):
    
    def get_happiness(self):
        
        data = collections.defaultdict(lambda: {})
        
        for line in self.input:
            subject, line = line.split(' would ')
            happiness, object_ = line.split(' happiness units by sitting next to ')
            object_ = object_[:-1]
            
            happiness = eval(happiness.replace('gain ', '').replace('lose ', '-'))
            
            # print(f'{subject=}, {happiness=}, {object_=}, {data=}')
            data[subject][object_] = happiness
            
        return data
    
    def calc_optimal_seating(self, include_me=False):
        
        happiness = self.get_happiness()
        people    = list(happiness.keys())
        
        if include_me:
            for person in people:
                happiness['me'][person] = 0
                happiness[person]['me'] = 0
            
            people    = list(happiness.keys())
        
        data = {}
        for seating in itertools.permutations(people, len(people)):

            total = 0
            for index in range(len(seating)):
                
                if index == 0:
                    total += happiness[seating[0]][seating[-1]]
                else:
                    total += happiness[seating[index]][seating[index - 1]]
                    
                if index == len(seating) - 1:
                    total += happiness[seating[index]][seating[0]]
                else:
                    total += happiness[seating[index]][seating[index + 1]]
            
            data[seating] = total
            
            # print(f'{seating=}, {total=}')
            
        series = pd.Series(data=data).sort_values()
        return series
    
    
input = '''Alice would gain 54 happiness units by sitting next to Bob.
Alice would lose 79 happiness units by sitting next to Carol.
Alice would lose 2 happiness units by sitting next to David.
Bob would gain 83 happiness units by sitting next to Alice.
Bob would lose 7 happiness units by sitting next to Carol.
Bob would lose 63 happiness units by sitting next to David.
Carol would lose 62 happiness units by sitting next to Alice.
Carol would gain 60 happiness units by sitting next to Bob.
Carol would gain 55 happiness units by sitting next to David.
David would gain 46 happiness units by sitting next to Alice.
David would lose 7 happiness units by sitting next to Bob.
David would gain 41 happiness units by sitting next to Carol.'''

assert Day13(input=input.split('\n')).calc_optimal_seating().iloc[-1] == 330

input = '''Alice would lose 2 happiness units by sitting next to Bob.
Alice would lose 62 happiness units by sitting next to Carol.
Alice would gain 65 happiness units by sitting next to David.
Alice would gain 21 happiness units by sitting next to Eric.
Alice would lose 81 happiness units by sitting next to Frank.
Alice would lose 4 happiness units by sitting next to George.
Alice would lose 80 happiness units by sitting next to Mallory.
Bob would gain 93 happiness units by sitting next to Alice.
Bob would gain 19 happiness units by sitting next to Carol.
Bob would gain 5 happiness units by sitting next to David.
Bob would gain 49 happiness units by sitting next to Eric.
Bob would gain 68 happiness units by sitting next to Frank.
Bob would gain 23 happiness units by sitting next to George.
Bob would gain 29 happiness units by sitting next to Mallory.
Carol would lose 54 happiness units by sitting next to Alice.
Carol would lose 70 happiness units by sitting next to Bob.
Carol would lose 37 happiness units by sitting next to David.
Carol would lose 46 happiness units by sitting next to Eric.
Carol would gain 33 happiness units by sitting next to Frank.
Carol would lose 35 happiness units by sitting next to George.
Carol would gain 10 happiness units by sitting next to Mallory.
David would gain 43 happiness units by sitting next to Alice.
David would lose 96 happiness units by sitting next to Bob.
David would lose 53 happiness units by sitting next to Carol.
David would lose 30 happiness units by sitting next to Eric.
David would lose 12 happiness units by sitting next to Frank.
David would gain 75 happiness units by sitting next to George.
David would lose 20 happiness units by sitting next to Mallory.
Eric would gain 8 happiness units by sitting next to Alice.
Eric would lose 89 happiness units by sitting next to Bob.
Eric would lose 69 happiness units by sitting next to Carol.
Eric would lose 34 happiness units by sitting next to David.
Eric would gain 95 happiness units by sitting next to Frank.
Eric would gain 34 happiness units by sitting next to George.
Eric would lose 99 happiness units by sitting next to Mallory.
Frank would lose 97 happiness units by sitting next to Alice.
Frank would gain 6 happiness units by sitting next to Bob.
Frank would lose 9 happiness units by sitting next to Carol.
Frank would gain 56 happiness units by sitting next to David.
Frank would lose 17 happiness units by sitting next to Eric.
Frank would gain 18 happiness units by sitting next to George.
Frank would lose 56 happiness units by sitting next to Mallory.
George would gain 45 happiness units by sitting next to Alice.
George would gain 76 happiness units by sitting next to Bob.
George would gain 63 happiness units by sitting next to Carol.
George would gain 54 happiness units by sitting next to David.
George would gain 54 happiness units by sitting next to Eric.
George would gain 30 happiness units by sitting next to Frank.
George would gain 7 happiness units by sitting next to Mallory.
Mallory would gain 31 happiness units by sitting next to Alice.
Mallory would lose 32 happiness units by sitting next to Bob.
Mallory would gain 95 happiness units by sitting next to Carol.
Mallory would gain 91 happiness units by sitting next to David.
Mallory would lose 66 happiness units by sitting next to Eric.
Mallory would lose 75 happiness units by sitting next to Frank.
Mallory would lose 99 happiness units by sitting next to George.'''

Day13(input=input.split('\n')).calc_optimal_seating().iloc[-1]
Day13(input=input.split('\n')).calc_optimal_seating(include_me=True).iloc[-1]
```

```python
class Day14(AdventSolution):
    
    def get_profiles(self):
        
        data = {}
        
        for line in self.input:
            reindeer, line = line.split(' can fly ')
            speed, line    = line.split(' km/s for ')
            duration, line = line.split(' seconds, but then must rest for ')
            rest, line     = line.split(' seconds.')
            
            data[reindeer] = {
                'speed':    int(speed),
                'duration': int(duration),
                'rest':     int(rest)}
            
        return pd.DataFrame(data=data).T
    
    def fly(self, reindeer, speed, duration, rest, until):
        
        distance = 0
        for time in range(until):
            time = time % (duration + rest)
            if time < duration:
                distance += speed
        return distance
    
    def race(self, until):
        
        df = self.get_profiles()
        
        return pd.Series({
            row.Index: self.fly(
                reindeer = row.Index,
                speed    = row.speed,
                duration = row.duration,
                rest     = row.rest,
                until    = until)
            for row
            in df.itertuples()}).sort_values()
    
input = '''Comet can fly 14 km/s for 10 seconds, but then must rest for 127 seconds.
Dancer can fly 16 km/s for 11 seconds, but then must rest for 162 seconds.'''

assert Day14(input=input.split('\n')).race(1000).iloc[-1] == 1120

input = '''Vixen can fly 19 km/s for 7 seconds, but then must rest for 124 seconds.
Rudolph can fly 3 km/s for 15 seconds, but then must rest for 28 seconds.
Donner can fly 19 km/s for 9 seconds, but then must rest for 164 seconds.
Blitzen can fly 19 km/s for 9 seconds, but then must rest for 158 seconds.
Comet can fly 13 km/s for 7 seconds, but then must rest for 82 seconds.
Cupid can fly 25 km/s for 6 seconds, but then must rest for 145 seconds.
Dasher can fly 14 km/s for 3 seconds, but then must rest for 38 seconds.
Dancer can fly 3 km/s for 16 seconds, but then must rest for 37 seconds.
Prancer can fly 25 km/s for 6 seconds, but then must rest for 143 seconds.'''

Day14(input=input.split('\n')).race(2503).iloc[-1]

```
