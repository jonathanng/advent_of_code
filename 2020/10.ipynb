```python
train_1 = '''16
10
15
5
1
11
7
19
6
12
4'''

train_2 = '''28
33
18
42
31
14
46
20
48
47
24
23
49
45
19
38
39
11
1
32
25
35
8
17
7
9
4
2
34
10
3'''

test = '''73
114
100
122
10
141
89
70
134
2
116
30
123
81
104
42
142
26
15
92
56
60
3
151
11
129
167
76
18
78
32
110
8
119
164
143
87
4
9
107
130
19
52
84
55
69
71
83
165
72
156
41
40
1
61
158
27
31
155
25
93
166
59
108
98
149
124
65
77
88
46
14
64
39
140
95
113
54
66
137
101
22
82
21
131
109
45
150
94
36
20
33
49
146
157
99
7
53
161
115
127
152
128'''

class JoltageFixer:
    
    def __init__(self, adapters):
        self.series = pd.Series(data=adapters.split('\n')).astype(int)
        extras      = pd.Series([0, self.series.max() + 3])
        self.series = self.series.append(extras).sort_values().reset_index(drop=True)
        
    def find_chain_distribution(self):
        return (self.series - self.series.shift(1)).fillna(self.series.min()).astype(int).value_counts()
    
    @functools.lru_cache()
    def get_arrangement_count(self, index=0):
        
        value = self.series.loc[index]

        if index == self.series.index[-1]:
            return 1

        next_index = index + 1
        last_value = -1

        _coumt = 0

        while True:

            if next_index > self.series.index[-1]:
                break

            next_value = self.series.loc[next_index]

            if next_value == last_value:
                break

            if next_value <= value + 3:
                _coumt += self.get_arrangement_count(index=next_index)

            else:
                break

            next_index += 1
            last_value  = next_value

        return _coumt
        
    
self = JoltageFixer(adapters=train_1)
chain_distribution = self.find_chain_distribution()
assert chain_distribution[1] == 7
assert chain_distribution[3] == 5
assert self.get_arrangement_count() == 8

self = JoltageFixer(adapters=train_2)
chain_distribution = self.find_chain_distribution()
assert chain_distribution[1] == 22
assert chain_distribution[3] == 10
assert self.get_arrangement_count() == 19208

self = JoltageFixer(adapters=test)
chain_distribution = self.find_chain_distribution()
chain_distribution[1] * chain_distribution[3]
self.get_arrangement_count() 
```
