```python
class ExpenseProcessor:
    
    def __init__(self, expenses):
        self.expenses = pd.Series(expenses).sort_values().reset_index(drop=True)
        
    def find_pair_product(self, _sum=2020):
        
        head_index = 0
        tail_index = self.expenses.index[-1]
        
        while head_index < tail_index:

            head = self.expenses.iloc[head_index]
            tail = self.expenses.iloc[tail_index]
            
            if head + tail == _sum:
                return head * tail

            elif head + tail > _sum:
                tail_index -= 1

            elif head + tail < _sum:

                head_index += 1

                while self.expenses.iloc[head_index] + self.expenses.iloc[tail_index] < _sum:
                    tail_index += 1
                    assert tail_index <= self.expenses.index[-1]
                    
    def find_triplet_product(self, _sum=2020):
        
        head_index = 0
        tail_index = self.expenses.index[-1]
        
        for head_index in self.expenses.index:
            for tail_index in reversed(self.expenses.index):
                
                head = self.expenses.iloc[head_index]
                tail = self.expenses.iloc[tail_index]

                if head + tail >= _sum:
                    continue
                    
                elif head_index < tail_index:
                    break

                else:

                    if _sum - head - tail in self.expenses.values:
                        return head * tail * (_sum - head - tail)
                    else:
                        continue


                    
        raise ValueError
                    
expenses = [
    1721,
    979,
    366,
    299,
    675,
    1456]

self = ExpenseProcessor(expenses=expenses)

assert self.find_pair_product(2020)    == 514579
assert self.find_triplet_product(2020) == 241861950
```
