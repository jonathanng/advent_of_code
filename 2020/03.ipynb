```python
class Toboggan:
    
    def __init__(self, _map):
        self.map = pd.DataFrame(data=[
            list(line)
            for line
            in _map.split('\n')])
        
    def get_map_x(self, x):
        max_col = self.map.columns.max()
        return ((x) % (max_col + 1))
    
    def get_path(self, x_delta=3, y_delta=1):
        
        x = 0
        y = 0
        
        path = []
        
        while y <= self.map.index.max():
            
            path.append(self.map.iloc[y, self.get_map_x(x=x)])
            
            x += x_delta
            y += y_delta
            
        return path
    
    def get_tree_count(self, x_delta=3, y_delta=1):
        
        path = self.get_path(
            x_delta = x_delta,
            y_delta = y_delta)
        
        return pd.Series(path).value_counts()['#'] 
        
self = Toboggan(_map=train)
assert self.get_tree_count() == 7

self = Toboggan(_map=test)
assert self.get_tree_count() == 169
```
