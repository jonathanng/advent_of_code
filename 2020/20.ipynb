```python
class ImageReader:
    
    def __init__(self, tiles):
        
        tile_id = None
        data    = collections.defaultdict(lambda: [])
        
        for line in tiles.split('\n'):
            if len(line) == 0:
                continue
            elif line.startswith('Tile '):
                tile_id = int(line[5:-1])
            else:
                data[tile_id].append(list(line))
                
        self.tiles = {
            tile_id: Tile(
                tile_id = tile_id,
                df      = pd.DataFrame(tile))
            for tile_id, tile
            in data.items()}
        
        self.remaining_tiles = sorted(list(self.tiles.keys()))
        self.tile_map        = {}
        
        self.place_tile(
            tile_id = self.remaining_tiles[0],
            x       = 0,
            y       = 0)
        
        while len(self.remaining_tiles) > 0:
            _ = self.place_next_tile()

        series = pd.Series(self.tile_map)
        series.index.names = ['x', 'y']
        self.tile_map = series.unstack().T
        
    def place_tile(self, tile_id, x, y):
        
        self.tile_map[x,y] = tile_id
        self.remaining_tiles.remove(tile_id)
        
    def place_next_tile(self):

        for (x, y), tile_id in self.tile_map.items():
            for (target_x, target_y, direction, edge) in [
                (x + 1, y, '→', self.tiles[tile_id].edge_right),
                (x - 1, y, '←', self.tiles[tile_id].edge_left),
                (x, y + 1, '↑', self.tiles[tile_id].edge_up),
                (x, y - 1, '↓', self.tiles[tile_id].edge_down)]:


                if (target_x, target_y) in self.tile_map.keys():
                    continue

                possible_tiles = []                
                for possible_tile_id in self.remaining_tiles:

                    possible_tile = self.tiles[possible_tile_id]
                    
                    assert possible_tile_id not in self.tile_map.values()
                    assert possible_tile.tile_id not in self.tile_map.values()

                    possible_tiles += [
                        (possible_tile, direction, edge, matching_direction, mirror)
                        for matching_direction, mirror
                        in possible_tile.find_edge_matches(edge=edge)]

                if len(possible_tiles) == 1:

                    matching_tile, direction, edge, matching_direction, mirror = possible_tiles[0]

                    matching_tile.rotate(
                        direction          = direction,
                        edge               = edge,
                        matching_direction = matching_direction,
                        mirror             = mirror)
                    
                    self.place_tile(
                        tile_id = matching_tile.tile_id,
                        x       = target_x,
                        y       = target_y)

                    return self
                
    def get_image(self):
        df = pd.concat([
            pd.concat([self.tiles[tile_id].df.iloc[1:-1, 1:-1] for tile_id in row], axis=1)
            for _, row
            in self.tile_map[::-1].iterrows()], axis=0).T

        return pd.DataFrame(np.rot90(df, k=3)).T
    
class Tile:

    def __init__(self, tile_id, df):
        
        self.tile_id = tile_id
        self.df      = df        
        self.size    = len(df.index)
    
    @property
    def edge_up(self):
        return ''.join(self.df.iloc[0].values)

    @property
    def edge_down(self):
        return ''.join(self.df.iloc[self.size - 1].values)[::-1]

    @property
    def edge_left(self):
        return ''.join(self.df[0].values)[::-1]
    
    @property
    def edge_right(self):
        return ''.join(self.df[self.size - 1].values)
    
    def find_edge_matches(self, edge):
        
        matches = []

        if edge == self.edge_up[::-1]:
            matches.append(('↓', None))
        if edge == self.edge_up:
            matches.append(('↓', '⇄'))
            
        if edge == self.edge_down[::-1]:
            matches.append(('↑', None))
        if edge == self.edge_down:
            matches.append(('↑', '⇄'))
            
        if edge == self.edge_left[::-1]:
            matches.append(('→', None))
        if edge == self.edge_left:
            matches.append(('→', '⇅'))            
            
        if edge == self.edge_right[::-1]:
            matches.append(('←', None))
        if edge == self.edge_right:
            matches.append(('←', '⇅'))
            
        return matches
    
    def rotate(self, direction, edge, matching_direction, mirror):
        
        rotation = {
            ('←', '←'): 0,
            ('↑', '↑'): 0,
            ('→', '→'): 0,
            ('↓', '↓'): 0,
            
            ('←', '↓'): 90,
            ('↑', '←'): 90,
            ('→', '↑'): 90,
            ('↓', '→'): 90,
            
            ('←', '→'): 180,
            ('↑', '↓'): 180,
            ('→', '←'): 180,
            ('↓', '↑'): 180,

            ('←', '↑'): 270,
            ('↑', '→'): 270,
            ('→', '↓'): 270,
            ('↓', '←'): 270}[(direction, matching_direction)]

        if mirror is None:
            pass
        elif mirror == '⇄':
            self.df = self.df.T.sort_index(ascending=False).reset_index(drop=True).T
        elif mirror == '⇅':
            self.df = self.df.sort_index(ascending=False).reset_index(drop=True)
        
        for _ in range(int(rotation / 90)):
            self.df = pd.DataFrame(np.rot90(self.df.values, k=3))
            
        assert {
            '←': self.edge_right,
            '↑': self.edge_down,
            '→': self.edge_left,
            '↓': self.edge_up,
        }[direction][::-1] == edge
        
        return self
    
    def __repr__(self):
        return f'Tile {self.tile_id}'
    
self = ImageReader(tiles=train)
assert np.product([
    self.tile_map.iloc[0,0],
    self.tile_map.iloc[0,-1],
    self.tile_map.iloc[-1,0],
    self.tile_map.iloc[-1,-1]]) == 20899048083289

self = ImageReader(tiles=test)
np.product([
    self.tile_map.iloc[0,0],
    self.tile_map.iloc[0,-1],
    self.tile_map.iloc[-1,0],
    self.tile_map.iloc[-1,-1]])
```

```python
import tqdm

MONSTER = '''                  # 
#    ##    ##    ###
 #  #  #  #  #  #   '''

class MonsterFinder:
    
    def __init__(self, tiles):
        
        self.image = ImageReader(tiles=tiles).get_image()
        
        self.monster = pd.DataFrame(data=[
            list(line)
            for line
            in MONSTER.split('\n')])
        
        self.is_monster = collections.defaultdict(lambda: False)
        
        for monster in self.monster_iter():
            for x in tqdm.tqdm(range(len(self.image) - len(monster.columns))):
                for y in range(len(self.image.index) - len(monster.index)):

                    image = self.image.iloc[y:y + len(monster.index), x:x + len(monster.columns)]

                    self.find_monster(
                        monster = monster,
                        image   = image)
        
    def monster_iter(self):

        for mirror in [True, False]:
            for k in range(4):
                monster = np.rot90(self.monster, k=k)
                if mirror:
                    monster = np.fliplr(monster)

                yield pd.DataFrame(monster)
                
    def find_monster(self, monster, image):

        assert monster.shape == image.shape

        for y_delta, row in monster.iterrows():

            image_row = image.iloc[y_delta]

            df = pd.DataFrame({
                'monster': row.values,
                'image':   image_row.values})

            if ((df['monster'] == '#') & (df['image'] != '#')).sum() != 0:
                return
            
        stack = monster.stack()
        for y_delta, x_delta in stack[stack == '#'].index:
            self.is_monster[x_delta + image.columns[0], y_delta + image.index[0]] = True
        
self = MonsterFinder(tiles=train)
assert (self.image == '#').sum().sum() - len(self.is_monster) == 273

self = MonsterFinder(tiles=test)
(self.image == '#').sum().sum() - len(self.is_monster)
```
