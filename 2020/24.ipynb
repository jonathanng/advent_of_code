```python
class Floor:
    
    def __init__(self, flipped_tiles):
        
        self.flip_counts = collections.defaultdict(lambda: 0)
        
        for tile in flipped_tiles.split('\n'):
            x, y = self.get_tile(tile)
            self.flip_counts[x, y] += 1
            
    def get_tile(self, tile):
        
        x = 0
        y = 0
        
        while len(tile) > 0:
            
            is_y_even = abs(y) % 2 == 0
            
            if tile.startswith('e'):
                neighbor = 'e'
                x += 1
            elif tile.startswith('se'):
                neighbor = 'se'
                y -= 1
                if is_y_even:
                    x += 1
            elif tile.startswith('sw'):
                neighbor = 'sw'
                y -= 1
                if not is_y_even:
                    x -= 1
            elif tile.startswith('w'):
                neighbor = 'w'
                x -= 1
            elif tile.startswith('nw'):
                neighbor = 'nw'
                y += 1
                if not is_y_even:
                    x -= 1
            elif tile.startswith('ne'):
                neighbor = 'ne'
                y += 1
                if is_y_even:
                    x += 1
            else:
                raise NotImplementedError

            tile = tile[len(neighbor):]
                
        return x, y
    
    def get_board(self, is_bool=False):
        df = pd.Series(self.flip_counts, name='flip_count').unstack().fillna(0).T.sort_index(ascending=False)
        if is_bool:
            df = df.applymap(lambda x: '◼︎' if x % 2 == 1 else '◻︎')
        return df
    
    def neighbor_iter(self, x, y):

        is_y_even = abs(y) % 2 == 0
        
        for _y in [y - 1, y, y + 1]:
            
            for _x in [x - 1, x, x + 1]:
                
                if (_x, _y) == (x, y):
                    continue
                    
                if is_y_even:
                    if ((_x, _y) == (x - 1, y + 1)
                     or (_x, _y) == (x - 1, y - 1)):
                        continue
                else:
                    if ((_x, _y) == (x + 1, y + 1)
                     or (_x, _y) == (x + 1, y - 1)):
                        continue
                    
                yield _x, _y
                
    def wait_day(self):
        
        board = self.get_board()
        
        new_flip_counts = collections.defaultdict(lambda: 0)
        new_flip_counts.update(self.flip_counts)
        
        for x in range(board.columns.min() - 1, board.columns.max() + 2):
            for y in range(board.index.min() - 1, board.index.max() + 2):

                black_count = sum([
                    self.flip_counts[_x, _y] % 2 == 1
                    for _x, _y
                    in self.neighbor_iter(x, y)])
                
                if self.flip_counts[x, y] % 2 == 0:
                    # Any white tile with exactly 2 black tiles immediately adjacent to it is flipped to black.
                    if black_count == 2:
                        new_flip_counts[x, y] += 1
                else:
                    # Any black tile with zero or more than 2 black tiles immediately adjacent to it is flipped to white.
                    if ((black_count == 0)
                    or (black_count > 2)):
                        new_flip_counts[x, y] += 1
                    
        self.flip_counts = new_flip_counts
        
        return self


self = Floor(flipped_tiles=train)
assert self.get_tile('se') == (1, -1)
assert self.get_tile('sw') == (0, -1)
assert self.get_tile('ne') == (1, 1)
assert self.get_tile('nw') == (0, 1)
assert self.get_board().applymap(lambda x: x == 1).sum().sum() == 10
assert self.get_board().applymap(lambda x: x == 2).sum().sum() == 5

for _ in range(100):
    _ = self.wait_day()
assert self.get_board().applymap(lambda x: x % 2 == 1).sum().sum() == 2208

self = Floor(flipped_tiles=test)
self.get_board().applymap(lambda x: x % 2 == 1).sum().sum()

for _ in range(100):
    _ = self.wait_day()
self.get_board().applymap(lambda x: x % 2 == 1).sum().sum()
```
